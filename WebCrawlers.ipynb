```python
def get_article_urls(url, pattern):
    result = []
    
    try:
        req = urllib.request.Request(url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        soup = BeautifulSoup(webpage)

        for a_tag in soup.find_all('a', href=True):
            if re.match(pattern, a_tag["href"]):
                result.append(a_tag["href"])
                
    except Exception as e:
        print("Unable to access= "+url)
        print(e)
    
    return result
```

```python
def get_article(url):
    try:
        req = urllib.request.Request(url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        soup = BeautifulSoup(webpage)
        text = ' '.join(map(lambda x: x.getText(), soup.select('p')))
        
    except Exception as e:
        print("Unable to access= "+url)
        print(e)
        text = ""
        
    return text
```

```python
import re
from bs4 import BeautifulSoup
import urllib.request
```

```python
# First crawler

seed_url = "https://www.federalreserve.gov/newsevents/pressreleases.htm"
website_name = "https://www.federalreserve.gov"
result = []
news_list_urls = []
seen = set()

req_main = urllib.request.Request(seed_url,headers={'User-Agent': 'Mozilla/5.0'})
webpage_main = urllib.request.urlopen(req_main).read()
soup = BeautifulSoup(webpage_main)

for a_tag in soup.find_all('a', href=True):
    if a_tag["href"][:26] == "/newsevents/pressreleases/":
        news_list_urls.append(website_name+a_tag["href"])

while len(result) < 10:
    article_urls = get_article_urls(news_list_urls.pop(0), "/newsevents/pressreleases/")
    
    for article_url in article_urls:
        full_article_url = website_name+article_url
        if full_article_url in seen:
            continue
            
        print("Scraping", full_article_url)
        article = get_article(full_article_url).lower()
        if re.search("covid", article):
            result.append(website_name+article_url)
            
        seen.add(full_article_url)
        print("Found", len(result), "articles")
        if len(result) >= 10:
            break
```

```python
print(result)
```

```python
# Second Crawler
import re
from bs4 import BeautifulSoup
import urllib.request

seed_url = "https://www.sec.gov/news/pressreleases"
website_name = "https://www.sec.gov"
result = []
seen = set()

article_urls = get_article_urls(seed_url, "/news/press-release")

while len(result) < 20:
    article_url = article_urls.pop(0)
    full_article_url = website_name+article_url
    if full_article_url in seen:
        continue

    print("Scraping", full_article_url)
    article = get_article(full_article_url)
    if re.search("charges", article.lower()):
        result.append((website_name+article_url, article))

    seen.add(full_article_url)
    print("Found", len(result), "articles")
```
